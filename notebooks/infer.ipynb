```python
import sys
sys.path.append('../')
from blood_segmentation.generator import RetinalBloodVesselGenerator as Generator
from PIL import Image
from matplotlib import pyplot as plt
```

```python
# Load generator
gen = Generator(net_type='unet', pretrained_model='../models/unet.pth')
```

```python
# Load image
filename = '00944016_fundus_4.png'
image = Image.open('../sample/' + filename)
image = image.convert(mode='RGB') # for gray image
```

```python
# Segment blood vessel and plot
output = gen.generate(image) # may take a while without gpu
```

```python
fig = plt.figure(figsize=[10,10])
plt.subplot(1,2,1), plt.imshow(image), plt.axis('off')
plt.subplot(1,2,2), plt.imshow(output), plt.axis('off')
```
